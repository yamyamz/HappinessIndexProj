```python
%matplotlib inline
my_plt.close("all")
```

```python
df_row.plot(x ='Year', y='Happiness Score', kind = 'line')
my_plt.show()
```

From this line graph, I can conclude that the happiness score of Singapore decreased from 2016 to 2021.

```python
df_row_line = df_row.loc[:][['Year','Generosity','Social Support from family and friends','Health (Life Expectancy)','Freedom','Trust (Government Corruption)']]
plot = df_row_line.plot(kind='line', x='Year',figsize=(12,6))

my_plt.xlabel("Year", fontsize = 12)
my_plt.title("Trend of change in the indicators",fontsize=18)
```

And from the graph above we can see the trend of change in the 5 indicatos

```python
df_row_bivariate = df_row.loc[:][['Year','Happiness Score','Generosity','Social Support from family and friends','Health (Life Expectancy)','Freedom','Trust (Government Corruption)']]
plot = df_row_bivariate.plot(kind ='barh', x ='Year' , figsize = (12,8))

my_plt.xlabel("amount", fontsize = 12)
my_plt.title("Change in Gen and",fontsize=18)
```

The graph above shows the trend in happiness score and the 5 indicators side by side.

```python
df_row_bivariate = df_row.loc[:][['Happiness Score','Generosity','Social Support from family and friends','Health (Life Expectancy)','Freedom','Trust (Government Corruption)']]

df = df_row_bivariate.astype(float)
ib_corr = df.corr()

ib_corr
```

<h2>Step 8: Drawing Insights </h2>

<h2>Step 9: Finding out if the external factors like unemployment and crime cases have an impact on the happiness score of Singapore.</h2>

1) Draw graphs to see trend
2) Formulate a correlation table
3) Represent the correlation in a heat map using seaborn

```python
#Step 1 Drawing graph to see the trend in crime cases and unemployment 

#line
df_2_line = df_2.loc[:][['Year','crime case','unemployment']]
df = df_2_line.astype(float)
plot = df.plot(kind='line', x='Year',figsize=(12,6))

my_plt.xlabel("Year", fontsize = 12)
my_plt.title("Trend of change in the external Factors",fontsize=18)


#bar
df_2_bivariate = df_2.loc[:][['Year','crime case','unemployment']]
df = df_2_bivariate.astype(float)

plot = df.plot(kind ='barh', x ='Year' , figsize = (12,8))

my_plt.xlabel("Year", fontsize = 12)
my_plt.title("Change in Gen and",fontsize=18)
```

This graph shows that the both the crime cases and the unemployment increased from 2016 to 2020.

```python
df_2_bivariate = df_2.loc[:][['Happiness Score','crime case','unemployment']]

df = df_2_bivariate.astype(float)
tb_corr = df.corr()

tb_corr
```

With that, I calcuted the correlation between the external factors (crime cases and happiness score) and the happiness score. I displayed this correlation in a heat map below, using seaborn.

<h2>Step 10: Plotting a heat map</h2>
Here I chose to use seaborn instead of matplotlib as matplotlib is mainly used for basic plotting like bars, lines and scatter plots. Whereas for seaborn, it provides a variety of visualisation patterns and uses fewer syntax. It also helps us to do statistics visualisation.


```python
my_plt.figure(figsize=(6,4))
sns.heatmap(tb_corr, annot= True , cbar = True)
```

<h2> Step 11: In conclusion</h2>

The correlation between the happiness score and the crime case is -0.52, they have a strong negative linear correlation whereby when the happiness score decrease, the crime case increased. Although it is not in one of the indicators, it still refects a strong negative correlation ( better correlation than the indicators themselves)

The correlation between the happiness score and the unemployment is -0.37 and its means they have a strong negatice linear correlation (not as strong os the crime case though as indicated by the lighter shade of colour in the heat map). So when the unemployment rate increased, the happiness score decreased, as it should. And this refects a stronger negative correlation than the other indicators.

My conclusion is yes, these external factors do play a part in determining the happiness score although they might not be one of the 6 indicators.
